```python
df= pd.read_csv("customer_shopping_behavior.csv")
```

```python
#checking the missing values 
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column [feature engineering] -- age group 
labels = ['Young', 'Adult', 'Middel_aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used', axis = 1)
```

```python
import pandas as pd
from sqlalchemy import create_engine

server = "localhost,1433"
database = "customer_behavior"
driver = "ODBC Driver 17 for SQL Server"

engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}"
    f"?driver={driver}&trusted_connection=yes"
)

# Write DataFrame to SQL Server
df.to_sql(
    name="customer",
    con=engine,
    if_exists="replace",
    index=False
)

print("✅ Data successfully inserted into SQL Server")

# Read data back
df_check = pd.read_sql("SELECT TOP 5 * FROM customer", engine)
print(df_check)
```
